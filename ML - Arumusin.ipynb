```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import random
import pickle
import time
```

```python
DIRECTORY = r'C:\Users\sarum\ADS_Project_FiMeSi\ADS_Projekt_FMS\Pictures'
CATEGORIES = ['Cats','Dogs']
```

```python
# IMG_SIZE = 50 #< bereits erledigt oder? - NR. 3

data = [] # NR. 4

for category in CATEGORIES:
    folder = os.path.join(DIRECTORY, category)
    print(folder)
    label = CATEGORIES.index(category)
    for img in os.listdir(folder):
        img_path = os.path.join(folder, img)
        # print(img_path) #< zur Kontrolle NR. 1
        # break #< zur Kontrolle NR. 1
        img_array = cv2.imread(img_path)
        # img_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE)) #< bereits erledigt oder? - NR. 3
        data.append([img_array, label]) # NR. 4
        # plt.imshow(img_array) #< zur Kontrolle - NR. 2, evtl. (img_array, , cmap = 'gray')
        # break #< zur Kontrolle NR. 2
```

```python
len(data)
```

```python
random.shuffle(data)
```

```python
data[0]
```

```python
X = [] #features
y = [] #labels

for features, labels in data:
    X.append(features)
    y.append(labels)
```

```python
X = np.array(X)
y = np.array(y)
```

```python
pickle.dump(X, open('X.pkl', 'wb'))
pickle.dump(y, open('y.pkl', 'wb'))      
```

```python
X = pickle.load(open('X.pkl', 'rb'))
y = pickle.load(open('y.pkl', 'rb'))
```

```python
X = X/255
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
model = Sequential()

model.add(Conv2D(64, (3,3), activation = 'relu'))
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(64, (3,3), activation = 'relu'))
model.add(MaxPooling2D((2,2)))

model.add(Flatten())

model.add(Dense(128, input_shape = X.shape[1:], activation = 'relu'))

model.add(Dense(2, activation = 'softmax'))
```

```python
model.compile(optimizer = 'adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=5, validation_split=0.1)
```
