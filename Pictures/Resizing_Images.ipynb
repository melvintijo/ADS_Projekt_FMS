```python
from PIL import Image
import glob
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
img_path = r"C:\Users\thekkenin\OneDrive - ZHAW\Master\Semester 2\Applied Data Science\Projekt\Pictures\Dogs\11648797464.jpg"

im = Image.open(img_path)
print('{}'.format(im.format))
print('size: {}'.format(im.size))
print('image mode: {}'.format(im.mode))
im.show
```

```python
#epty list
image_dog_list = []
resized_dog_images = []
```

```python
#append images to list
for filename in glob.glob(r"C:\Users\thekkenin\OneDrive - ZHAW\Master\Semester 2\Applied Data Science\Projekt\Pictures\Dogs\*.jpg"):
    print(filename)
    img = Image.open(filename)
    image_dog_list.append(img)
```

```python
#append resized images to list
for image in image_dog_list:
    image = image.resize((500,500))
    resized_dog_images.append(image)
```

```python
# save resized images to new folder
for (i,new) in enumerate(resized_dog_images):
    new.save('{}{}{}'.format(r'C:\Users\thekkenin\OneDrive - ZHAW\Master\Semester 2\Applied Data Science\Projekt\Pictures\Dogs\Resized_images\dogs' , i+1,'.jpg'))
```
